```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
initial_val = np.random.randint(0,10)

x_coef = 0.6
eps_coef = -0.4

def gen_time_series(N, current_vals, x_coef, eps_coef, n=0):

    if n==N:

        return current_vals

    else:

        new_val = current_vals[-1] + np.random.normal()
        current_vals.append(new_val)

        return gen_time_series(N, current_vals, x_coef, eps_coef, n+1)

test = gen_time_series(2500,[initial_val], x_coef, eps_coef)

plt.plot(test)
```
